### Решение


$$
J = ||X - AS||_F^2 = tr((X - AS)(X - AS)^T) = \\[6pt]
tr[(X - AS)(X - AS)^T = (X - AS)(X^T - S^TA^T)] = tr(XX^T) - tr(ASX^T) - tr(XS^TA^T) + tr(ASS^TA^T)
$$

Заметим что
$$
tr(ASX^T) = tr(X^TAS) = tr(SX^TA) \\[6pt]
tr(XS^TA^T) = tr((XS^TA^T)^T) = tr(ASX^T) = tr(SX^TA)
$$

Поэтому
$$
J = tr(XX^T) - 2tr(SX^TA) + tr(ASS^TA^T) \\[6pt]
\frac{dJ}{dS} = \frac{tr(XX^T)}{dS} - \frac{2tr(SX^TA)}{dS} + \frac{tr(ASS^TA^T)}{dS} \\[6pt]
$$

$$
\frac{d[tr(XX^T)]}{dS} = 0 \\[6pt]
\frac{d[tr(SX^TA)]}{dS} = \text{[Using ex.2]} = (X^TA)^T = A^TX  \\[6pt]
\frac{d[tr(ASS^TA^T)]}{dS} = \frac{d[tr(S^TA^TAS)]}{dS} = \bigg[\text{пусть $A^TA = Z$}\bigg] = \frac{d[tr(S^TZS)]}{dS} \\[6pt]
$$


$$
[S^TZ]_{k,l} = \sum_r [S^T]_{k,r} [Z]_{r,l} = \sum_r [S]_{r,k} [Z]_{r,l} \\[6pt]
[S^TZS]_{k,j} = \sum_l [S^TZ]_{k,l} [S]_{l,j} = \sum_l (\sum_r [S]_{r, k} [Z]_{r,l}) [S]_{l,j} \\[6pt]
tr(S^TZS) = \sum_i [S^TZS]_{i,i} = \sum_i \sum_l \sum_r [S]_{r,i} [Z]_{r,l} [S]_{l,i} \\[6pt]
$$

$$
[\frac{tr(S^TZS)}{dS}]_{a,b} = \bigg[\text{дает вклад только если r=a, i=b или l=a, i=b}\bigg] = \sum_l [Z]_{a,l} [S]_{l,b} + \sum_r [S]_{r,b} [Z]_{r,a} = \\[6pt]
= \bigg[\text{$[Z]_{r,a} = [Z]_{a,r}$}\bigg] = \sum_l [Z]_{a,l} [S]_{l,b} + \sum_r [Z]_{a,r} [S]_{r,b}  = 2 \sum_l [Z]_{a,l} [S]_{l,b} \rArr 
\frac{tr(S^TZS)}{dS} = 2ZS = 2A^TAS
$$

Итого:

$$
\frac{dJ}{dS} = 2A^TAS + A^TX = 2A^T(AS - X)
$$

```python
def j(X, A, S):
    return np.linalg.norm(X - A @ S, 'fro')**2

def grad_j_S(X, A, S):
    return 2 * A.T @ (A @ S - X)
```
